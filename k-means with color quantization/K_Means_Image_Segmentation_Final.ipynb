```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from sklearn.cluster import KMeans
```

```python
img = cv2.imread('data.jpg')
cv2_imshow(img)
```

```python
img2 = img.reshape((-1,3))
img2 = np.float32(img2)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(img2)
```

```python
for i in range(0,kmeans.labels_.size):
  if kmeans.labels_[i] == 0:
    img2[i] = np.array([0.0,0.0,255.0])
  elif kmeans.labels_[i] == 1:
    img2[i] = np.array([255.0,255.0,0.0])
  elif kmeans.labels_[i] == 2:
    img2[i] = np.array([0.0,0.0,0.0])
  elif kmeans.labels_[i] == 3:
    img2[i] = np.array([255.0,0.0,255.0])
```

```python
res2 = img2.reshape((img.shape))
cv2.imwrite("answer.jpg" , res2)
cv2_imshow(res2)
```
